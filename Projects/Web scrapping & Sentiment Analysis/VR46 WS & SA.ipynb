```python
#cleaning and preprocessing article1 
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
#Tokenization
v_token=word_tokenize(v)
print(v_token)
```

```python
#Check the legth of the article
len(v_token)
```

```python
#Frequency of words
from nltk.probability import FreqDist
fdist=FreqDist()
```

```python
for word in v_token:
    fdist[word.lower()]+=1
fdist
```

```python
fdist_top=fdist.most_common(15)
fdist_top
```

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
v_stem=ps.stem(v)
```

```python
v_sc=re.sub('[^a-zA-Z.]',' ',v_stem)
print(v_sc)
```

```python
v_sw= ' '.join([word for word in v_sc.split() if word not in (stopwords.words('english'))])
print(v_sw)
```

```python
for token in v_token:
    pos_tag=nltk.pos_tag([token])
    #print(pos_tag)
```

```python
#Import required libraries 
from wordcloud import WordCloud
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
wc=WordCloud(width=800,height=800,background_color='white',min_font_size=10).generate(v_sw) 
```

```python
#Ploting the WordCloud image(Most repeated words appear big)                       
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wc) 
plt.axis("off") 
plt.tight_layout(pad = 0)
plt.show() 
```

```python
#SENTIMENT ANALYZER
from nltk.sentiment.vader import SentimentIntensityAnalyzer
         
sia = SentimentIntensityAnalyzer()
ss = sia.polarity_scores(v_sw)
print(ss)
```

```python
if ss["compound"]>=0.5:
    print("positive")
elif ss["compound"]<=-0.5:
    print("negative")
else:
    print("neutral")
```

```python
#sSentiment percentage in pie chart
labels=['negative','neutral','positive']
sizes=[ss['neg'],ss['neu'],ss['pos']]
plt.pie(sizes,labels=labels,autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```
